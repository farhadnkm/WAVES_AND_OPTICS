```python
import numpy as np
from matplotlib import pyplot as plt
```

# definition of a ray
A ray is line that can be defined by a vector.
Rays can have origins.

```python
src = np.array([1, 0])
ray = np.array([0, 1]) # +x direction
end = src+ray

x = np.array([src[1], end[1]])
y = np.array([src[0], end[0]])

plt.plot(x, y, marker='o')
plt.show()
```

# definition of many rays
Here, a broad source (in 2D, AKA line source) is defined. The rays are defined by unit vectors with different angles.

```python
N = 100
length = 10
src_x = np.zeros((N), np.float32)                   # position of sources
src_y = np.linspace(-0.5, 0.5, N)

ray_angles = np.linspace(-30, 30, N) * np.pi/180    # from -30 degrees to 30 degrees with N steps
ray_x = length * np.cos(ray_angles)                 # now we can find the Cartesian components
ray_y = length * np.sin(ray_angles)


# drawing the rays
fig, ax = plt.subplots(figsize=[6, 6])
for i in range(N):
    x = [src_x[i], src_x[i]+ray_x[i]]
    y = [src_y[i], src_y[i]+ray_y[i]]
    ax.plot(x, y, color='blue', linewidth=0.5, alpha=0.5)

# drawing the sources
ax.scatter(src_x, src_y)

ax.set_xlim(-0.2, 1.8)
ax.set_ylim(-1, 1)
plt.show()
```
